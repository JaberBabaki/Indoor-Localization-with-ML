```python
import pandas as pd
import tensorflow as tf
import numpy as np
```

```python
df = pd.read_csv('BLE_rssi_V5.csv')
df.sample(frac=1)
```

```python
# Import train_test_split function
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
import collections


X = df[['beacon' + str(i) for i in range(1, 6)]]
Y = df['location']

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3) # 70% training and 30% test


# flag = 0
# while flag == 0:
#     dict_classifications = {}
#     for i in range(1, 100):
#         #Create KNN Classifier
#         knn = KNeighborsClassifier(n_neighbors=i)

#         #Train the model using the training sets
#         knn.fit(X_train, y_train)

#         #Predict the response for test dataset
#         y_pred = knn.predict(X_test)
#         # Model Evaluation for k=5



#         # Model Accuracy, how often is the classifier correct?
#         acc = metrics.accuracy_score(y_test, y_pred)
#         if acc not in dict_classifications:
#             dict_classifications[acc] = [y_test, y_pred]

#     od = collections.OrderedDict(sorted(dict_classifications.items()))
#     if list(od.keys())[0] > 0.63:
#         print('accurancy : ',list(od.keys())[0])
#         print('*****************************************************')
#         print(od[list(od.keys())[0]])

        

        
dict_classifications = {}
for i in range(1, 100):
    #Create KNN Classifier
    knn = KNeighborsClassifier(n_neighbors=i)

    #Train the model using the training sets
    knn.fit(X_train, y_train)

    #Predict the response for test dataset
    y_pred = knn.predict(X_test)
    # Model Evaluation for k=5

    # Model Accuracy, how often is the classifier correct?
    acc = metrics.accuracy_score(y_test, y_pred)
    if acc not in dict_classifications:
        dict_classifications[acc] = [y_test, y_pred]

od = collections.OrderedDict(sorted(dict_classifications.items()))

print(list(od.keys())[0])
print(od[list(od.keys())[0]])
```
